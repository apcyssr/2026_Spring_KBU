```python
# --- ส่วนที่ 1: ติดตั้งแบบบังคับเวอร์ชัน ---
!apt-get update
!apt-get install -y chromium-browser chromium-chromedriver
!pip install selenium

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# --- ส่วนที่ 2: ตั้งค่าเพื่อข้ามปัญหา Session ---
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
# บังคับให้ใช้พื้นที่ชั่วคราว /tmp เพื่อเลี่ยงปัญหาสิทธิ์การเข้าถึง (ที่อาจติดจากชื่อภาษาเกาหลี)
chrome_options.add_argument('--user-data-dir=/tmp/user-data')

try:
    print("⏳ กำลังเริ่มระบบ Browser...")
    driver = webdriver.Chrome(options=chrome_options)

    # ลองเข้าหน้าเว็บตูนตามโจทย์
    driver.get("https://comic.naver.com/webtoon")
    time.sleep(3)

    print(f"✅ สำเร็จ! กำลังดึงข้อมูลจาก: {driver.title}")

    # ดึงชื่อเรื่องการ์ตูนมาโชว์ (เพื่อยืนยันว่าทำการบ้านได้)
    titles = driver.find_elements(By.CSS_SELECTOR, "span[class*='ContentTitle__title']")
    print(f"\n📌 พบการ์ตูน {len(titles)} เรื่อง:")
    for t in titles[:5]: # โชว์ 5 เรื่องแรก
        print(f"- {t.text}")

except Exception as e:
    print(f"❌ ยังมีปัญหา: {e}")

finally:
    if 'driver' in locals():
        driver.quit()
```
